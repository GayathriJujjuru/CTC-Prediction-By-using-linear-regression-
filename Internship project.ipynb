```python
x=dataa[['College_T1','College_T2','Role_Manager','City_Metro','previous CTC','previous job changes','Graduation marks','Exp']]
y=dataa['Actual CTC']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.5,test_size=0.3,random_state=100)
```

```python

lr =LinearRegression()
lr.fit(x_train,y_train)
```

```python
prediction=lr.predict(x)
dataa['Predicted CTC']= prediction
dataa
```

```python
dataa['Predicted CTC']
```

```python
import seaborn as sns
sns.lmplot(x='Actual CTC',y='Predicted CTC',data=dataa,scatter_kws={'color':'blue'},line_kws={'color':'red'})
plt.show()
```

```python
from sklearn import metrics
print('Mean square value:',metrics.mean_squared_error(y,prediction))
```

```python
print('Root mean square value:',np.sqrt(metrics.mean_squared_error(y,prediction)))
```
